```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sys
import io
import plotly.express as px
import plotly.graph_objects as go 
```

```python
df = pd.read_csv(r'C:\Users\pardis\Desktop\digikala\orders.csv')

df.head()
```

```python
df_location = df['city_name_fa']
orders = df_location.value_counts()
orders
```

```python
type(orders)
```

```python

orders.head()
cityes = orders.index.tolist()[:10]
amount = orders.values.tolist()[:10]
```

```python
fig = px.bar (x = cityes , y = amount , title = 'توزیع سفارشات بر اساس شهر' ,labels={'x':'شهر','y':'تعداد سفارش'})
fig.show()
```

```python

fig = px.scatter(x = cityes , y = amount  ,title = 'توزیع سفارشات بر اساس شهر' ,labels={'x':'شهر','y':'تعداد سفارش'})
fig.show()
```
